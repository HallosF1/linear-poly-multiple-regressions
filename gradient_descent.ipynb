```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(df, learning_rate=0.1, cur_x = 3, precision = 0.000001, previous_step_size = 1, max_iters = 10000):
    """
    learning_rate: default 0.01
    cur_x: start point
    precision: stop point
    df: function
    """
    iters = 0 

    while previous_step_size > precision and iters < max_iters:
        prev_x = cur_x
        cur_x = cur_x - learning_rate * df(prev_x)
        previous_step_size = abs(cur_x - prev_x)
        iters = iters+1
        if iters == 5000:
            learning_rate = 0.01
        print("Iteration",iters,"\nX value is",cur_x)

    print("The global minimum occurs at", cur_x)
```

```python
df = lambda x: x ** 2
gradient_descent(df, max_iters=10000)
```
